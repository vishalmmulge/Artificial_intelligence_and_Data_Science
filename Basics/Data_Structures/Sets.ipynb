```python
##create a set
my_set = {1,2,3,4,5,6}
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set()
print(type(my_empty_set))
```

```python
my_empty_set = set([1,2,3,4,5,6,7])
print(type(my_empty_set))
print(my_empty_set)
```

```python
my_set = set([1,2,3,4,5,6,7,2,4])
print(type(my_set))
print(my_set)
```

```python
#basic set operations
##ading and removing
my_set.add(8)
print(my_set)
my_set.remove(2)
print(my_set)

```

```python
my_set.remove(9)
```

```python
my_set.discard(11)
print(my_set)
```

```python
##pop method
removed_element = my_set.pop()
print(removed_element)
print(my_set)
```

```python
##clear all the elements
my_set.clear()
print(my_set)
```

```python
##Set membership test
my_set = set([1, 2, 3, 4, 5, 6, 7, 2, 4])
print(2 in my_set)
print(10 in my_set)
```

```python
##Mathematical Operations
set1 = {1,2,3,4,5}
set2 = {4,5,6,7,8}

##union
union_set = set1.union(set2)
print(union_set)

##Intersection
intersection_set = set1.intersection(set2)
print(intersection_set)

##Difference
diff_set = set1.difference(set2)
print(diff_set)

##Symmetric Difference
sym_diff_set = set1.symmetric_difference(set2)
print(sym_diff_set)
```
