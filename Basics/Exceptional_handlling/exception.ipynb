```python
### try, except, else, finally block

try:
    num1 = int(input("Enter the first number: "))
    num2 = int(input("Enter the second number: "))
    result = num1 / num2
except ValueError as ex:
    print(ex)
    print("Please enter valid integers.")
except ZeroDivisionError as ex:
    print(ex)
    print("Please check the denominator value, it should not be zero.")
except Exception as ex1:
    print(ex1)
    print("Main exception block")
else:
    print(f"The result is: {result}")
finally:
    print("Execution completed.")
```

```python
## try, except, else, finally block with function
def divide_numbers():
    try:
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))
        result = num1 / num2
    except ValueError as ex:
        print(ex)
        print("Please enter valid integers.")
    except ZeroDivisionError as ex:
        print(ex)
        print("Please check the denominator value, it should not be zero.")
    except Exception as ex1:
        print(ex1)
        print("Main exception block")
    else:
        print(f"The result is: {result}")
    finally:
        print("Execution completed.")


divide_numbers()
```
