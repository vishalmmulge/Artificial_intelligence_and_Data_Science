```python
##Checking if the path is a file or directory

import os

path = 'example.txt'
if os.path.isfile(path):
    print(f"file exists in this path '{path}'")
elif os.path.isdir(path):
    print(f"directory exists in this path '{path}'")
else:
    print(f"'{path}' is niether file nor directory")
```
