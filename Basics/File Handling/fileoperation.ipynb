```python
### Reading a whole file

with open('example.txt',mode='r') as file:
    content = file.read()
    print(content)
```

```python
### Reading line by line

with open("example.txt",mode='r') as file:
    for line in file:
        print(line.strip()) ### .strip() removes the newline char
```

```python
### Writing a file (OverWriting)

with open('example.txt',mode='w') as file:
    file.write("hey whos there \n")
    file.write("Knock knock")
```

```python
### write a file without overwritting

with open('example.txt', 'a') as file:
    file.write("\n Append operation taking place \n")
    file.write("Knock knock again\n")
```

```python
### writing a list of lines for a file
lines = ["1st line \n", "2nd line\n" , "3rd line\n"]
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
##Binary Files

# Writing toa binary file
data = b'\x00\x01\x02\x03'
with open('example.bin', 'wb') as file:
    file.write(data)
```

```python
with open('example.bin','rb') as file:
    content = file.read()
    print(content)
```

```python
### Read the content from sorce text file n write to a destination text file
with open('example.txt','r') as file:
    content1 = file.read()

with open('destination.txt', 'w') as file:
        file.write(content1)
```

The W+ mode in Python is used to open a file for both reading and writing. If the file does not exist. it will be created. If the file exist, it is truncated that is the file is overwritten

```python
with open('example.txt','w+') as file:
    file.write("Hello worlds\n")
    file.write("this is a new line")

    ## Move the file pointer to the beginning to read the content
    file.seek(0)

    ## Read the content of the file
    with open('example.txt', 'r') as file:
        content = file.read()
    print(content)
```
