```python
def Odd_or_even(num):
    """This function finds even or odd"""
    if num%2==0:
        print(f"{num} is a even number")
    else:
        print(f"{num} is a odd number")
```

```python
##Call this function
Odd_or_even(24)
```

```python
## function with multiple parameters

def add(a,b):
    sum = a + b
    return sum
result = add(5,6)
print(f"{result} is sum of A and B")
```

```python
## Default Parameter
def greet(name, greeting="Hello"):
    message = f"{greeting}, {name}!"
    return message

print(greet("Alice","Good morning"))
print(greet("Bob"))
```

```python
##Variable length arguments
## Positional arguments

def print_num(*arg):
    for num in arg:
        print(num)
```

```python
print_num(1,2,3,4,5,6,7,8,9,"Vishal")
```

```python
## keywords arguments
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(
    name="Alice",
    age=30,
    city="New York")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional arguments{val}")
        
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python

print_details(1, 2, 3, name="Alice", age=30)
```
