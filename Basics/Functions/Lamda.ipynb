```python
#syntax 
lambda arguments : expression
```

```python
def add(a,b):
    return a + b
```

```python
add(2,3)
```

```python
addition=lambda a,b:a+b
print(type(addition))
print(addition(5,5))
```

```python
def even(num):
    if num%2 == 0:
        return True
    else:
        return False

even(5)
```

```python
even1 = lambda num : num%2 == 0
even1(2)
```

```python
def add(x,y,z):
    return x+y+z
add(5,5,5)
```

```python
add1 = lambda x,y,z : x+y+z
add1(5, 5, 15)
```

```python
## map()
num = [1,2,3,5,6]
squared = list(map(lambda x: x**2, num))
print(squared)
```
